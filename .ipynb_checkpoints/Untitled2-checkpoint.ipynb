```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def get_line(X_data, Y_data) :
    
    w = tf.Variable(5.0)
    X = tf.placeholder(tf.float32)
    Y = tf.placeholder(tf.float32)
    
    hypothesis = w * X
    
    cost = tf.reduce_sum(tf.square(hypothesis - Y))
    
    # Minimize
    learning_rate = 0.00001
    gradient = tf.reduce_mean(((w * X) - Y) * X)
    descent = w - learning_rate * gradient
    update = w.assign(descent)
    
    #gvs = optimizer.compute_gradients(cost, [w])
    #apply_gradient = optimizer.apply_gradients(gvs)
    
    sess = tf.Session()
    sess.run(tf.global_variables_initializer())
    
    for step in range(120):
        sess.run(update, feed_dict={X : X_data, Y : Y_data })
        print(step,  sess.run(w))
        
    return sess.run(w)
```

```python
lemon = [97, 98, 110, 134, 159, 103, 143, 123, 134, 140, 162, 130, 109, 122, 98, 81, 115, 131, 122, 71, + \
        83, 112, 120, 121, 156, 176, 104, 96, 100, 88, 76]
orange = [67, 67, 77, 99, 118, 69, 101, 86, 95, 98, 120, 95, 75, 85, 62, 50, 76, 92, 85, 42, 50, 75, + \
         82, 82, 113, 129, 68, 63, 66, 57, 47]
temperature = [70, 72, 71, 76, 78, 82, 81, 82, 80, 82, 83, 84, 77, 78, 75, 74, 77, 81, 78, 70, 77, 80, + \
              81, 82, 84, 83, 80, 82, 81, 82, 82]
leaflets = [90, 90, 104, 98, 135, 90, 135, 113, 126, 131, 135, 99, 99, 113, 108, 90, 126, 122, 113, 0, 90, + \
            108, 117, 117, 135, 158, 99, 90, 95, 81, 68]


# Set benefits
benefits = []
for i in range(31) :
    if i < 14 :
        price = 0.25
        
    elif 14 <= i and i < 25 : 
        price = 0.5
    
    else :
        price = 0.35
        
    benefits.append((lemon[i] + orange[i]) * price)

lemon_orange = []
for i in range(31) :
    lemon_orange.append(lemon[i] + orange[i])
    

plt.plot(range(0, 31), benefits)
plt.show()
#get_line(lemon_orange, benefits)
```

```python
avg_salesVolume_aboutTemp = []
temps = []

for temp in range(60, 160):
    sum = 0
    for i in range(0, 31):
        
        if leaflets[i] == temp:
            sum += lemon_orange[i]
            
    if leaflets.count(temp) != 0:
        temps.append(temp)
        avg_salesVolume_aboutTemp.append(sum / leaflets.count(temp))
                                         
                                
plt.plot(temps, avg_salesVolume_aboutTemp)
domain = range(60, 160)
codomain = [1.8443725 * i for i in domain]
plt.plot(domain, codomain)
plt.show()
```
